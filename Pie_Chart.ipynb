```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nbins = 5
totalN = 100
```

```python
rawdata = np.ceil(np.logspace(np.log10(1/2), np.log10(nbins-0.1), totalN))

uniquenums = np.unique(rawdata)
data4pie = np.zeros(len(uniquenums))

for i in range(len(uniquenums)):
  data4pie[i] = sum(rawdata == uniquenums[i])
```

```python
plt.pie(data4pie, labels=100*data4pie/sum(data4pie))
plt.show()
```

```python
plt.pie(data4pie, labels=['zero', 'one', 'two', 'three', 'four'], explode=[0,0,0,0.3,0])
plt.show()
```
